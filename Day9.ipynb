```python
import numpy as np
import copy
ini = np.zeros((1000, 1000), dtype=int)

p = [500, 500]
f = open("./data/day9_input.txt", "rt")

lines = f.readlines()
f.close() 

lNode = []
lNode.append(p)
for i in range(9):
    lNode.append(copy.deepcopy(p))



for line in lines:
    d, n = line.split(" ")
    n = int(n)
    lPP = []

    for i in range(n):
        lPP.append(copy.deepcopy(lNode[0]))
        if d == 'R':
            lNode[0][1] += 1
        elif d == 'L':
            lNode[0][1] -= 1
        elif d == 'U':
            lNode[0][0] -= 1
        else:
            lNode[0][0] += 1
            
             
        for k in range(1,10):
            if (abs(lNode[k][0] - lNode[k - 1][0]) + abs(lNode[k][1] - lNode[k - 1][1]) > 2):
                if (lNode[k][0] > lNode[k - 1][0]):
                    lNode[k][0] -= 1
                else:
                    lNode[k][0] += 1
                if (lNode[k][1] > lNode[k - 1][1]):
                    lNode[k][1] -= 1
                else:
                    lNode[k][1] += 1    
                    
                ini[lNode[9][0]][lNode[9][1]] = 1
            elif (abs(lNode[k][0] - lNode[k - 1][0]) > 1):
                lNode[k][1] = lNode[k - 1][1]
                if (lNode[k][0] > lNode[k - 1][0]):
                    lNode[k][0] = lNode[k - 1][0] + 1
                else:
                    lNode[k][0] = lNode[k - 1][0] - 1
                
                ini[lNode[9][0]][lNode[9][1]] = 1
            elif (abs(lNode[k][1] - lNode[k - 1][1]) > 1):
                lNode[k][0] = lNode[k - 1][0]
                if (lNode[k][1] > lNode[k - 1][1]):
                    lNode[k][1] = lNode[k - 1][1] + 1
                else:
                    lNode[k][1] = lNode[k - 1][1] - 1
                    
                ini[lNode[9][0]][lNode[9][1]] = 1
            
            else:
                break

print(sum(sum(ini)))
```
