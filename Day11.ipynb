## part1

```python
def ops(start, o1, o2):
    if (o2 == "old"):
        o2 = start
    else:
        o2 = int(o2)
    
    if (o1 == "+"):
        return start + o2
    elif (o1 == "-"):
        return start - o2
    elif (o1 == "*"):
        return start * o2
    else:
        return start / o2
```

```python
f = open("./data/day11_input.txt", "rt")
lines = f.readlines()
f.close() 

number = int (len(lines) / 7) + 1
l4all = []

for i in range(number):
    l4all.append([])
    
o1l = []
o2l = []
divisiblel = []
throwTl = []
throwFl = []
insl = []

for i in range(number):
    line1 = lines[7 * i + 1].strip().split(": ")[-1].split(", ")
    starting = [eval(t) for t in line1]
    o1, o2 = lines[7 * i + 2].strip().split("= old ")[-1].split(" ")
    divisible = int (lines[7 * i + 3].strip().split("by ")[-1])
    throwT = int (lines[7 * i + 4].strip().split(" ")[-1])
    throwF = int (lines[7 * i + 5].strip().split(" ")[-1])
    
    o1l.append(o1)
    o2l.append(o2)
    divisiblel.append(divisible)
    throwTl.append(throwT)
    throwFl.append(throwF)
    insl.append(0)
    
    for start in starting + l4all[i]:
        insl[i] += 1
        worry = int (ops(start, o1, o2) / 3)
        if (worry % divisible == 0):
            l4all[throwT].append(worry)
        else:
            l4all[throwF].append(worry)
    l4all[i].clear()

for r in range(1, 20):
    for i in range(number):
        for start in l4all[i]:
            insl[i] += 1
            worry = int (ops(start, o1l[i], o2l[i]) / 3)
            if (worry % divisiblel[i] == 0):
                l4all[throwTl[i]].append(worry)
            else:
                l4all[throwFl[i]].append(worry)   
        l4all[i].clear()

insl.sort()
print(insl[-1] * insl[-2])
```

## part2

```python
from math import lcm

number = int (len(lines) / 7) + 1
l4all = []

for i in range(number):
    l4all.append([])
    
o1l = []
o2l = []
divisiblel = []
throwTl = []
throwFl = []
insl = []

for i in range(number):
    line1 = lines[7 * i + 1].strip().split(": ")[-1].split(", ")
    starting = [eval(t) for t in line1]
    o1, o2 = lines[7 * i + 2].strip().split("= old ")[-1].split(" ")
    divisible = int (lines[7 * i + 3].strip().split("by ")[-1])
    throwT = int (lines[7 * i + 4].strip().split(" ")[-1])
    throwF = int (lines[7 * i + 5].strip().split(" ")[-1])
    
    o1l.append(o1)
    o2l.append(o2)
    divisiblel.append(divisible)
    throwTl.append(throwT)
    throwFl.append(throwF)
    insl.append(0)
    
    for start in starting + l4all[i]:
        insl[i] += 1
        worry = ops(start, o1, o2)
        if (worry % divisible == 0):
            l4all[throwT].append(worry)
        else:
            l4all[throwF].append(worry)       
    l4all[i].clear()

# to aviod the number increase so quick, so take the mod of lcm
lcmN = lcm(divisiblel[0], divisiblel[1])
for i in range(2, number):
    lcmN = lcm(lcmN, divisiblel[i])

for r in range(1, 10000):
    for i in range(number):
        for start in l4all[i]:
            insl[i] += 1
            worry = ops(start, o1l[i], o2l[i]) % lcmN
            if (worry % divisiblel[i] == 0):
                l4all[throwTl[i]].append(worry)
            else:
                l4all[throwFl[i]].append(worry)
        l4all[i].clear()
    
insl.sort()
print(insl[-1] * insl[-2])
```
